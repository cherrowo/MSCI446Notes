Training Set: Trains model

<h3>Underfitting & Overfitting</h3>
When we split dataset, we might remove some important data that could help with training model, so we don't want to assign too much weight to test set.

If we overlearn our set the inferences might be too specific and not work for new information

<h4>Underfitting</h4>
Happens when model can't capture relationship between input features and target outputs, or underlying pattern. <b>High Bias.</b>
Cause: model too simple and we don't give enough training data

<h4>Overfitting</h4>
Model does a good job in training but performs poorly in test set. Sensitive to small fluctuations in training set. <b>High Variance</b>
Cause: model too complex, too many features



<h3>Bias-Variance Tradeoff</h3>
Bias: indication of model error or difference between forecast and true values
Variance: flexibility of the model and efforts to precisely capture each data point during training

impossible to have low bias and low variance. always a tradeoff between the two.

GOAL: create model that generalises well to new data, but can still capture underlying patterns in training

if underfitting, increase complexity to reduce bias. if overfitting, simplify model to reduce variance
